```python
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt

```

```python
# Step 1: Create the dataset
X = np.array([1, 2, 3, 4, 5, 6, 7, 8]).reshape(-1, 1) 
y = np.array([0, 0, 0, 0, 1, 1, 1, 1]) #Labels: fail/pass

```

```python
# Step 2: Train the logistic regression model
model = LogisticRegression()
model.fit(X, y)

```

```python
# Step 3: Predict probabilities
X_test = np.linspace(1, 8, 100).reshape(-1, 1)
y_prob = model.predict_proba(X_test)[:, 1]  # Probability of passing

```

```python
# Step 4: Plot the logistic curve
plt.plot(X_test, y_prob, color='blue', label='Probability of Passing',linestyle='--')
plt.scatter(X, y, color='black', label='Actual Data')
plt.xlabel('Hours Studied')
plt.ylabel('Pass Probability')
plt.title('Logistic Regression: Hours vs Pass')
plt.legend()
plt.grid(True)
plt.show()
```
