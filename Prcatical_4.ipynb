```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = np.array([

                 [2.5,2.4,1.2],
                 [0.5,0.7,0.3],
                 [2.2,2.9,1.4],
                 [1.9,2.2,1.1],
                 [3.1,3.0,1.7],
                 [2.3,2.7,1.5],
                 [2.0,1.6,0.9],
                 [1.0,1.1,0.4],
                 [1.5,1.6,0.8],
                 [1.1,0.9,0.2]

])
```

```python
df = pd.DataFrame(data,columns=['X','Y','Z'])
```

```python
STAND_DATA= (df - df.mean())/df.std()
```

```python
cov_matrix = np.cov(STAND_DATA.T)
```

```python
eig_vals, eig_vecs = np.linalg.eig(cov_matrix)
```

```python
sort_indices = np.argsort(eig_vals)[::-1]
eig_vals = eig_vals[sort_indices]
eig_vecs = eig_vecs[:, sort_indices]
```

```python
eig_vecs_subset = eig_vecs[:, :2]
```

```python
STAND_DATA.to_numpy().dot(eig_vecs_subset)
reduced_df = pd.DataFrame(STAND_DATA.to_numpy().dot(eig_vecs_subset), columns=['PC1', 'PC2'])
```

```python
print("\n Reduces Data (after _PCA)\n\n",reduced_df)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(reduced_df['PC1'],reduced_df['PC2'],color='blue',edgecolor='k')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Reduced Data (after PCA)')
plt.grid(True)
plt.show()
```
